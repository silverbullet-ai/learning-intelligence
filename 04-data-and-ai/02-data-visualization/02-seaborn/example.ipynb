```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
## scatter plot
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips)
plt.title('Scatter Plot for Total Bill vs Tip')
plt.show()
```

```python
## line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line Plot for Total Bill vs Size')
plt.show()
```

```python
## Categorical Plots
## Bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot for Day vs Total Bill')
plt.show()
```

```python
## Box plot
sns.boxplot(x='day',y='tip',data=tips)
plt.title('Finding Outliers')
plt.show()
```

```python
## Violin Plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Violin Plot')
plt.show()
```

```python
## Histogram

sns.histplot(tips['total_bill'],bins=10)
plt.title('Histogram withot Kernal Density Estimator(kde)')
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.title('Histogram with Kernal Density Estimator(kde)')
plt.show()
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
plt.title('KDE with fill')
plt.show()
```

```python
sns.kdeplot(tips['total_bill'])
plt.title('KDE without fill')
plt.show()
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
## Heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr,annot=True)
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd

df = pd.read_csv('../resources/sales_data.csv')
df.head()
```

```python
## Plot Total Sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product',y='Revenue',data=df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
## Plot Total Sales by region
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Revenue',data=df,estimator=sum, errorbar=None)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
