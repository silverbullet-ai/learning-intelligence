```python
## Reanaming columns
df = df.rename(columns={'Customer_Age':'Cust_Age'})
```

```python
df['Cust_Age'] = df['Cust_Age'].fillna(df['Cust_Age'].mean()).astype(int)
```

```python
df['New_Price'] = df['Unit_Price'].apply(lambda x:x+x*0.1)
```

```python
## Data Aggregating and Grouping
df.head()
```

```python
grouped_mean = df.groupby('Category')['Profit'].mean()
print(grouped_mean)
```

```python
product_group = df.groupby('Product')['Profit'].mean()
print(product_group)
```

```python
quantity_group = df.groupby('Product')['Quantity'].sum()
print(quantity_group)
```

```python
price_group = df.groupby('Product')['Unit_Price'].mean()
print(price_group)
```

```python
print(df['Unit_Price'].mean())
```

```python
grouped_sum = df.groupby(['Product','Region'])['Profit'].sum()
print(grouped_sum)
```

```python
## aggregate multiple functions
grouped_agg = df.groupby('Region')['Revenue'].agg(['mean','sum','count'])
grouped_agg
```

```python
### Merging and joining Dataframes
df1 = pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'], 'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='Key',how='inner')
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```
