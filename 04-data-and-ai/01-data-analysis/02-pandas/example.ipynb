# Pandas

```python
import pandas as pd
```

```python
## Series
data = [1,2,3,4,5]
series = pd.Series(data)
print(f"Series\n{series}")
print(type(series))
```

```python
## Creating Series from dictonary

dict_data = {'a':1, 'b':2, 'c':3}
series_dict = pd.Series(dict_data)
print(series_dict)
```

```python
nums = [1,2,3,4,5]
roman = ['i', 'ii', 'iii', 'iv', 'v']
pd.Series(roman,nums)
```

```python
## Dataframe

### Creating from Dictonary of Lists

data = {
    'Name':['Aahish', 'Manav'],
    'Age':[24,23],
    'City':['Gaya','Bengaluru']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
### Creating from List of Dictonaries

data = [
    {'Name':'Aahish','Age':24, 'City':'Gaya'},
    {'Name':'Manav', 'Age':23, 'City':'Bengaluru'}
]
df = pd.DataFrame(data)
print(df)
print(type(df))
```
