# Arithmetic Operators

```python
a = 10
b = 5

add_result = a + b              # addition operation
subtract_result = a - b         # subtraction operation
multiply_result = a * b         # multiplication operation
divide_result = a / b           # division operation
floor_divide_result = a // b    # floor division operation
modulus_result = a % b          # modulus operation
exponentiation_result = a ** b  # exponentiation operation

print(add_result)               # Output: 15
print(subtract_result)          # Output: 5
print(multiply_result)          # Output: 50
print(divide_result)            # Output: 2.0
print(floor_divide_result)      # Output: 2
print(modulus_result)           # Output: 0
print(exponentiation_result)    # Output: 100000
```

# Comparison Operators

```python
# Equal to operator (==)

a = 10
b = 10

a == b
```

```python
str1 = "Aahish"
str2 = "Aahish"

string1 = "Aahish"
string2 = "aahish"

print(str1 == str2)
print(string1 == string2)
```

```python
# Not equal to operator (!=)

print(str1 != str2)
print(string1 != string2)
```

```python
# Greater than operator (>)

num1 = 45
num2 = 55

print(num1 > num2) 
```

```python
# Less than operator (<)

num1 < num2
```

```python
# Greather then or equal to operator (>=)

number1 = 100
number2 = 100

n1 = 55
n2 = 45

print(number1 >= number2)
print(n1 >= n2)
```

```python
# Less then or equal to operator (<=)

print(number1 <= number2)
print(n1 <= n2)
```

# Logical Operators

```python
# And operator (and)

X = True
Y = True
Z = False

print(X and Y)  # Output: True
print(X and Z)  # Output: False
```

```python
# Or operator (or)

A = True
B = False
C = False

print(A or B)  # Output: True
print(B or C)  # Output: False
```

```python
# Not operator (not)

T = True
F = False

print(not T)  # Output: False
print(not F)  # Output: True
```
