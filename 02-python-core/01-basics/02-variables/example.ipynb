```python
## Declaring and assigning a variable

age = 23
name = 'Aahish'
height = 5.5
is_student = True
```

```python
## printing variables

print('age: ', age)
print('name: ', name)
print('height: ', height)
print('is_student: ', is_student)
```

```python
# Naming Conventions

# valid variable names

first_name = 'Aahish'  # snake_case
FirstName = 'Aahish'   # PascalCase
firstName = 'Aahish'   # camelCase
_lastname = 'Aayan'  # leading underscore
age1 = 23  # trailing number
```

```python
# invalid variable names

# 1name = 23  # starts with a number
# first-name = 'Aahish'  # contains a hyphen
# first name = 'Aahish'  # contains a space
# @name = 'Aahish'  # contains a special character
```

```python
# Dynamic Typing

age = 23 
name = 'Aahish'
height = 5.5
is_student = True

print(type(age))  # <class 'int'>
print(type(name))  # <class 'str'>
print(type(height))  # <class 'float'>
print(type(is_student))  # <class 'bool'>
```

```python
# Type Conversion(Casting)

age = 23
age_str = str(age)  # convert int to str
print(age_str, type(age_str))  # '23' <class 'str'>
```

```python
# Dynamic Typing

var = 10
print(var, type(var))  # 10 <class 'int'>

var = 'Hello'
print(var, type(var))  # 'Hello' <class 'str'>

var = 3.14
print(var, type(var))  # 3.14 <class 'float'>
```
