# while loop

```python
count = 0

while count < 5:
    print(count)
    count += 1
```

```python
count = 0

while count % 2 == 0:
    print(count)
    count += 1
```

# Loop Control Statements

## break

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

## continue

```python
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

# pass

```python
for i in range(5):
    if i == 3:
        pass
    print(i)
```

## nested loops

```python
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

### Examples

```python
# calculate the sum of 1st N natural number using for and while loop

# while loop

n = 10
sum = 0
count = 1

while count <= n:
    sum += count
    count += 1

print("Sum of first",n,"natural nubmers:",sum)
```

```python
# for loop

n = 12
sum = 0

for i in range(n+1):
    sum += i

print("Sum of first",n,"nautral numbers:",sum)
```

```python
## prime numbers between 1 and 100.

for num in range(1,101):
    if num > 1:
        for i in range(2,num):
            if num % i == 0:
                break
        else:
            print(num)
```
