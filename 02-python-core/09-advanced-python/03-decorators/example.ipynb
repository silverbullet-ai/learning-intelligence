```python
 ## Decorator
def master_welcome(func):
    def sub_welcome_method():
        print('Welcome to the world of Data Science')
        func()
        print('Here we play with data and numbers')
    return sub_welcome_method()
```

```python
@master_welcome
def intro():
    print('Silver Bullet Reporting!!!')
```

```python
## Decorator
def my_dec(fun):
    def wrapper():
        print('Before funcation call')
        fun()
        print('After function call')
    return wrapper
```

```python
@my_dec
def exhaust():
    print('V12 engine revs')
```

```python
exhaust()
```

```python
# Decorator with argument

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print('hello')
```

```python
say_hello()
```
