# Generators

```python
def square(n):
    for i in range(n):
        return i**2
```

```python
square(3)
```

```python
# with yield

def square(n):
    for i in range(n):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
a
```

```python
## Practical Example: Reading a large file line by line using generators

def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
```

```python
file_path = 'Revuelto.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
