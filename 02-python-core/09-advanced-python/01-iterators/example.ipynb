# Iterators

```python
my_list = [1, 2, 3, 4, 5]

for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
## Iterator

iterator = iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
## Iterate through elements

next(iterator)
```

```python
iterator1 = iter(my_list)
```

```python
try:
    print(next(iterator1))
except StopIteration as e:
    print(f'No more elements in list')
```
