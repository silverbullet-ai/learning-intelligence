# Inheritence

```python
## Single Inheritence
## Parent Class

class Car:

    def __init__(self,windows,doors,enginetype): # This is a constructor method that initializes the attributes of the Car class.
                                                 # Constructors are special methods that are called when an object is created from a class. They are used to initialize the attributes of the object.
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    
    def drive(self):
        print(f'The person will drive the {self.enginetype} car.')
```

```python
car1 = Car(4,5,'petrol')
car1.drive()
```

```python
##child class

class Tesla(Car):
    def __init__(self, windows, doors, enginetype,is_selfdriving):
        super().__init__(windows, doors, enginetype)
        self.is_selfdriving = is_selfdriving

    def self_driving(self):
        print(f'Tesla supports self driving: {self.is_selfdriving}')
```

```python
tesla1 = Tesla(4,5,'electric',True)
tesla1.self_driving()
```

```python
tesla1.drive()
```

```python
## Multipe Inheritence

## Base Class 1
class Animal:
    def __init__(self,name):
        self.name = name        # 'name' is local parameter variable, it is temporary.
                                # 'self.name' is instance varialbe, is stores the value recived from 'name'. 
    
    def speak(self):
        print('Subclasses must implement this method.')

## Base Class 2
class Pet:
    def __init__(self,owner):
        self.owner = owner

## Derived Class
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    
    def speak(self):
        return f'{self.name} says woof'

# create an object
dog = Dog('Spike','Aahish')
print(dog.speak())
print(f"Owner:{dog.owner}")
```
