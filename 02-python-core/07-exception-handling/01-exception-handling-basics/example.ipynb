# Exception Handling

```python
a = b
```

```python
## Try Except block

try:
    a = b
except:
    print('The variable is not assigned')
```

```python
## Printing the error message given by complier

try:
    a = b
except NameError as e:
    print(e)
```

```python
res = 1/0
```

```python
try:
    res = 1/0
except ZeroDivisionError as e:
    print(e)
    print('Denominator should not be 0.')
```

```python
try:
    res = 1/2
    a = b
except ZeroDivisionError as e:
    print(e)
```

```python
try:
    res = 1/2
    a = b
except ZeroDivisionError as e:
    print(e)
except Exception as e1:
    print(e1)
    print('Main exception is caught here.')
```

```python
try:
    num = int(input('Enter a nubmer:'))
    res = 10/num
except ValueError:
    print('This is not a valid number.')
except ZeroDivisionError:
    print('Denominator cannot be 0.')
except Exception as e:
    print(e)
```

```python
## try, except, else block

try:
    num = int(input('Enter a nubmer:'))
    res = 10/num
except ValueError:
    print('This is not a valid number.')
except ZeroDivisionError:
    print('Denominator cannot be 0.')
except Exception as e:
    print(e)
else:
    print(f'The result is {res}')
```

```python
## try, except, else and finally block

try:
    num = int(input('Enter a nubmer:'))
    res = 10/num
except ValueError:
    print('This is not a valid number.')
except ZeroDivisionError:
    print('Denominator cannot be 0.')
except Exception as e:
    print(e)
else:
    print(f'The result is {res}')  # else will execute only if there is no error.
finally:
    print('Execution Completed.')  # finally will execute regardless of error or no error.
```
