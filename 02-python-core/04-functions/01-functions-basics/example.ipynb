# Functions

```python
## creating a function

def evev_or_odd(num):
    """This functions finds out even or odd number"""
    if num%2 == 0:
        print('Even number')
    else:
        print('Odd number')
```

```python
## calling a function

evev_or_odd(23)
```

```python
## function with multiple parameters

def add(a,b):
    return a + b
```

```python
result = add(2,4)
print(result)
```

```python
## Default Parameter

def greet(name='Guest'):
    print(f'Hello {name}. Welcome to the hotel moon view.')

greet('Aahish')
```

```python
## Variable length arguments

def print_numbers(*aahish):
    for number in aahish:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,'Aahish')
```

```python
## positional arguments

def print_number(*args):       ## args is standard name.
    for numbers in args:
        print(numbers)
```

```python
print_number(1,2,3,4,5,'Aahish')
```

```python
## keyword arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f'{key}:{value}')
```

```python
print_details(name='Aahish',age=32,country='India')
```

```python
def print_all(*args, **kwargs):
    for val in args:
        print(f'Positional argument:{val}')
    
    for key,val in kwargs.items():
        print(f'{key}:{val}')
```

```python
print_all(1,2,3,4,5,'Aahish', name='Aahish', age=23, country='India')
```

```python
## return statement

def multiply(a,b):
    return a*b
```

```python
multiply(2,3)
```

```python
## return multiple parameters

def multiply(a,b):
    return a*b, a
```

```python
multiply(2,5)
```
