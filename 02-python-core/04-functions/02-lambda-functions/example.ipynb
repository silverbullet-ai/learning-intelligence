# Lambda function

```python
## normal function

def addition(a,b):
    return a + b
```

```python
addition(2,5)
```

```python
## lambda function

add = lambda x,y : x+y
print(type(add))

res = add(3,6)
print(res)
```

```python
## even function

def even(number):
    if number%2==0:
        return True
    else:
        return False

print(even(12))
print(even(13))
```

```python
## equivalent lambda function

even1 = lambda num : num%2==0

print(even1(28))
print(even1(23))
```
