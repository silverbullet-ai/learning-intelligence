# SQLite3

```python
import sqlite3
```

```python
## Connect to database
connection = sqlite3.connect('database/example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
## Creating the table in database

cursor.execute('''
create table if not exists employees(
               id Integer Primary Key,
               name Text Not Null,
               age Integer,
               dept Text
               )
''')

connection.commit()
```

```python
## Inserting the values in table

cursor.execute('Insert into employees(name,age,dept) values ("Aahish",23,"AI")')
connection.commit()
```

```python
cursor.execute('Insert into employees(name,age,dept) values ("Manav",24,"Software Development")')
connection.commit()
```

```python
cursor.execute('Insert into employees(name,age,dept) values ("Raj",22,"IT")')
connection.commit()
```

```python
## reading the data from table 

cursor.execute('select * from employees')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## Updating the data in table
cursor.execute('''
update employees
set age = 23
where name = "Raj"
''')

connection.commit()
```

```python
## Working with sales data
connection = sqlite3.connect('database/sales_data.db')
cursor = connection.cursor()

cursor.execute('''
create table if not exists sales(
               id Integer Primary Key,
               date Text Not Null,
               product Text Not Null,
               sales Integer,
               region Text
               )
''')

sales_data = [
    ('2024-05-12', 'Laptop', 100, 'East'),
    ('2024-05-23', 'Headphone', 200, 'South'),
    ('2024-07-28', 'Mobile', 150, 'North'),
    ('2024-08-01', 'Mouse', 120, 'West')
]

cursor.executemany('''
Insert into sales(date,product,sales,region)
                   values(?,?,?,?)
''',sales_data)

connection.commit()
```

```python
cursor.execute('select * from sales')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## close the connection

connection.close()
```
