# Dictionary

## Creating Dictionary

```python
empty_dict = {}
print(empty_dict)
print(type(empty_dict))
```

```python
new_dict = dict()
print(new_dict)
print(type(new_dict))
```

```python
student = {'name':'Aahish', 'age':23, 'grade':'A'}
print(student)
print(type(student))
```

```python
# Logical Error

students = {'name':'Aahish', 'age':23, 'name':'Manav'} # keys should be unique
print(students)
```

## Accessing Dictionary Elements

```python
student = {'name':'Aahish', 'age':23, 'grade':'A'}
print(student)
```

```python
print(student['grade'])
print(student['age'])
```

```python
## using get() method

print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', 'Not Available'))
```

## Modifying Dictionary Elements

```python
print(student)
```

```python
student['age']=24       # update value for the key
print(student)

student['address'] = 'India'    # add a new key and value
print(student)

del student['grade']        # delete key and value
print(student)
```

## Dictionary Methods

```python
print(student)
```

```python
keys = student.keys()   # give all the keys
print(keys)

values = student.values()   # give all the values
print(values)

items = student.items()     # give all the key-value pairs
print(items)
```

```python
## Shallow copy

student_copy = student  # this is not a shallow copy as the change in original will affect the copy also.
print(student)
print(student_copy)
```

```python
student['name'] = 'Manav'
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() # this is shallow copy as the change in original will not affect the copy.
print(student)
print(student_copy1)
```

```python
student['name'] = 'Krishna'
print(student)
print(student_copy1)
```

## Iterating over dictionaries

```python
student
```

```python
for keys in student.keys():
    print(keys)
```

```python
for values in student.values():
    print(values)
```

```python
width = max(len(key) for key in student)

for key, value in student.items():
    print(f'{key:>{width}} : {value}')
```

## Nested Dictionaries

```python
## creating

student_record = {
    'student1': {'name':'Aahish', 'age':23, 'grade':'A'},
    'student2': {'name':'Manav', 'age':22, 'grade':'A'},
}
print(student_record)
```

```python
## accessing

print(student_record['student1']['name'])
```

```python
## iterating
width = max(len(key) for key in student_record)

for student_id, student_info in student_record.items():
    print(f'{student_id:>{width}} : {student_info}')
    for key, value in student_info.items():
        print(f'{key:>{width}} -> {value}')
    print()
```

## Dictionary Comprehension

```python
squares = {x:x**2 for x in range(1,6)}
print(squares)
```

```python
evens = {x:x**2 for x in range(1,11) if x % 2 == 0}
print(evens)
```

## Practical Examples

```python
## use dictionary to count the frequency of numbers in a list

nubmers = [1,2,3,4,3,3,4,2,4,4]
frequency = {}

for nubmer in nubmers:
    if nubmer not in frequency:
        frequency[nubmer] = 1
    else:
        frequency[nubmer] += + 1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1 = {'a':1, 'b':2}
dict2 = {'c':3, 'd':4}

merged_dict = {**dict1, **dict2}
print(merged_dict)
```
