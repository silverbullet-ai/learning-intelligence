```python
tple = ()
print(type(tple))
```

```python
tple1 = tuple()
print(type(tple1))
```

```python
numbers = tuple([1,2,3,4,5])
numbers
```

```python
mixed_tuple = (1, 'Aahish', 3.14, True)
mixed_tuple
```

## Tuple Operations

```python
## Concatenation

print(numbers,mixed_tuple,sep='\n')

concatenation_tuple = numbers + mixed_tuple

print(concatenation_tuple)
```

```python
## Repetition

print(mixed_tuple * 3)

print(numbers * 3)
```

## Immutable Nature

```python
lst = [1,2,3,4,5]
print(lst)
```

```python
lst[1] = 'Aahish'
print(lst)
```

## Packing and Unpacking

```python
## packing

packed_tuple = 1, 'Aahish', 3.14
print(packed_tuple)
```

```python
## unpacking

a,b,c = packed_tuple
print(a,b,c,sep='\n')
```

```python
## unpacking with *

numbers = (1, 2, 3, 4, 5, 6)
first, *middle, last = numbers
print(first,middle,last,sep='\n')
```

## Nested Tuple

```python
## Nested List

lst = [[1,2,3], ['Aahish', 'Nexus'], [3.14, 2.71]]
lst[0][2]
lst[0][:2]
```

```python
lst = [[1,2,3], ['Aahish', 'Nexus'],(3.14, 2.71)]
lst[2][1]
```

```python
## Nested Tuple

nested_tuple = ((1, 2, 3), ('a', 'b', 'c'), (True, False))

print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating through a tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=' ')
    print()
```
