## Creating a Set

```python
my_set = set()
print(type(my_set))
```

```python
number_set = {1, 2, 3, 4, 5}
print(number_set)
```

```python
list_set = set([1, 2, 3, 4])
print(list_set)
```

```python
name_set = {'Aahish', 'Manav', 'Singh', 'Aahish', 'Raj', 'Krishna', 'Manav'}
print(name_set)
```

## Basic Set Operations

```python
name_set
```

```python
## Adding Elements

name_set.add('Ram')
print(name_set)

name_set.add('Ram')
print(name_set)
```

```python
## Removing Elements 'remove()'

name_set.remove('Raj')
print(name_set)
```

```python
## Set Membership test

num_set = {1, 2, 3, 4, 5}
print(3 in num_set)
print(10 in num_set)
```

## Mathematical Operations

```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

print(set1, set2, sep='\n')
```

```python
## Union
union_set = set1.union(set2)
print(union_set)
```

```python
## Intersection
intersection_set = set1.intersection(set2)
intersection_set
```

```python
set1.intersection_update(set2)
set1
```

```python
set3 = {1, 2, 3, 4, 5, 6}
set4 = {4, 5, 6, 7, 8, 9}

print(set3, set4, sep='\n')
```

```python
## Difference
print(set3.difference(set4))
print(set4.difference(set3))
```

```python
## Symmetric Difference

print(set3.symmetric_difference(set4))
```

## Set Methods

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

print(s1, s2, sep='\n')
```

```python
## is subset
s1.issubset(s2)
```

```python
## is superset
s1.issuperset(s2)
```

```python
### Counting Unique words in text

text = 'Hey i am Aahish and i am currently learning AI and i am 23 years old'
words = text.split()
print(words)
print(len(words))

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
