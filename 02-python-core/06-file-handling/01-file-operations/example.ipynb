```python
# Write to file
with open("sample_files/example.txt", "w") as file:

    file.write("Hello World\n")
    file.write("Python File Handling")
```

```python
# Read file
with open("sample_files/example.txt", "r") as file:

    content = file.read()

print(content)
```

```python
# Append example
with open("sample_files/example.txt", "a") as file:

    file.write("\nAppending new line")


# Read line by line
with open("sample_files/example.txt", "r") as file:

    for line in file:
        print(line.strip())
```

```python
# Write binary data
with open("sample_files/example.bin", "wb") as file:

    file.write(b"Hello Binary World")


# Read binary data
with open("sample_files/example.bin", "rb") as file:

    content = file.read()

print(content)
```

```python
# Copy file content

with open("sample_files/source.txt", "r") as source_file:

    content = source_file.read()

with open("sample_files/destination.txt", "w") as destination_file:

    destination_file.write(content)

print("File copied successfully")
```
